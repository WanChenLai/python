```python
#程式交易_財金三乙 A105230016 賴婉貞
import openpyxl
import pandas as pd  #pandas是python的excel
import os  #os:找檔案用
import datetime   #例如：計算兩天買賣隔幾天
#from IPython.display import display  #有顯示的問題時用的
```

```python
def print_full(x):    #先定義一個變數 print_full(x)
    df = pd.DataFrame(x)  #x原本是list 轉變格式成pandas
    pd.set_option('display.max_rows', 2000)   #橫的最多2000筆資料
    pd.set_option('display.max_columns', 20)   #直的資料最多20個(變數)
    display(df)
```

```python
excelname= "kd"   #excel名稱
sheetname = "kd"   #工作表名稱

wb = openpyxl.load_workbook(str(excelname)+".xlsx")  #使用openpyxl(Package)用load_workbook去抓現有的excel檔案，把工作表名稱改成字串的形式
ws = wb[str(sheetname)]   #讀對應的工作表名稱(一定要是字串)

dataset = []
    #創造矩陣
i = 1    #在現有的excel工作表，檢查每個row的第一個colum是不是空值，是的話不加入，有數值的話加入矩陣資料
while(ws.cell(row=i, column=1).value!=None):   #cell：excel簡寫
    row = []
    j = 1
    while(j<=5):  
        row.append(ws.cell(row=i, column=j).value)
        j = j +1
    dataset.append(row)
    i = i + 1
    #最後會形成一個二維的矩陣
```

```python
pd.DataFrame(dataset).head()
```

```python
trade =[[0 for x in range(11)] for y in range(len(dataset))]
trade[0][0] = "date"
trade[0][1] = "open"
trade[0][2] = "high"
trade[0][3] = "low"
trade[0][4] = "close"
trade[0][5] = "rsv"
trade[0][6] = "k"
trade[0][7] = "d"
trade[0][8] = "buy_day"
trade[0][9] = "cum_day"
trade[0][10] = "sell_day"
```

```python
for i in range(1, len(trade)):
    trade[i][0] = dataset[i][0]          #日期
    trade[i][1] = float(dataset[i][1])
    trade[i][2] = float(dataset[i][2])
    trade[i][3] = float(dataset[i][3])
    trade[i][4] = float(dataset[i][4])   #收盤價
```

```python
pd.DataFrame(trade).head()
```

```python
for i in range(9, len(trade)):  
    H=[]
    L=[]
    k=50
    d=50
    for j in range(0, 9): 
        A=trade[i-j][2]
        H.append(A)
        B=trade[i-j][3]
        L.append(B)
        j=j+1
    high=max(H)
    low=min(L)
    rsv=(trade[i][4]-low)/(high-low)*100
    k=k*2/3+rsv*1/3
    d=d*2/3+k*1/3
    trade[i][6] = k
    trade[i][7] = d
    trade[i][5]=rsv
    i=i+1

trade[9][6] = 50
trade[9][7] = 50
```

```python
pd.DataFrame(trade).head(20)
```

```python
for i in range(10, len(trade)):
    if(trade[i][6] > trade[i][7])and(trade[i-1][9]==0):  #買入
        trade[i][8] = 1  #買 BUY
        trade[i][9] = 1  #持有 HOLD
        trade[i][10] = 0  #賣 SELL
    elif(trade[i][6] < trade[i][7])and(trade[i-1][9]==1):   #賣出
        trade[i][8] = 0
        trade[i][9] = 0
        trade[i][10] = 1
    elif(trade[i-1][9]>0):
        trade[i][8] = 0
        trade[i][9] = 1
        trade[i][10] = 0
```
